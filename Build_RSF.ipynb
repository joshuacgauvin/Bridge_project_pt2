```python
"""
Collapse long-format (bridge-year) NBI panel data to ONE ROW PER BRIDGE,
producing a standard right-censored (event, time) target for sksurv's
RandomSurvivalForest.

sksurv's RandomSurvivalForest does NOT support counting-process / start-stop
interval data -- it wants exactly one (event, time) pair per subject. This
script implements that "Option A" structure:

- time  = bridge_age at the survey where the bridge FIRST hit "poor"
          (event == 1), OR bridge_age at the LAST observed survey if the
          bridge never transitioned (right-censored)
- event = 1 if the bridge ever transitioned to poor, 0 otherwise
- All other covariates (including time-varying ones like climate and
  maintenance) are taken from THAT SAME ROW -- i.e. the row at event,
  or the row at last observation if censored. This is the most defensible
  default: it reflects the bridge's state right before/at the outcome,
  rather than averaging conditions across decades of very different risk.

bridge_age is built purely from YEAR_BUILT_027 per your confirmation.
Reconstruction is captured separately via `ever_reconstructed` and
`years_since_reconstruction`, which are kept as ordinary covariates
(not used to redefine the time clock).
"""

import pandas as pd
import numpy as np

# ---- 1. Load your panel (bridge-year) data ----
# df = pd.read_csv("your_bridge_panel.csv")
df = pd.read_csv("massachusetts_bridges_rsf_ready.csv")  # update path

id_col = "STRUCTURE_NUMBER_008"
year_col = "SURVEY_YEAR"
age_col = "bridge_age"
age_flag_col = "bridge_age_negative_flag"
event_col = "event"

# ---- 2. Check bridge_age data quality before using it as the time axis ----
n_bad_age = (df[age_flag_col] == 1).sum() if age_flag_col in df.columns else (df[age_col] < 0).sum()
print(f"Rows with negative/bad bridge_age: {n_bad_age} of {len(df)}")
bad_age_structures = df.loc[df[age_col] < 0, id_col].unique()
if len(bad_age_structures) > 0:
    print(f"{len(bad_age_structures)} structures have at least one negative bridge_age row. "
          f"These likely have a bad YEAR_BUILT_027 value -- inspect/fix before modeling.")
    # Example structures to check by hand:
    print(bad_age_structures[:10])

# ---- 3. Sort within each bridge by survey year ----
df = df.sort_values([id_col, year_col]).reset_index(drop=True)

# ---- 4. For each bridge, find the row to keep: ----
#    - if the bridge ever has event == 1, keep the FIRST such row
#    - otherwise, keep the LAST observed row (most recent survey = censoring time)
df["_cum_event"] = df.groupby(id_col)[event_col].cummax()
df["_is_first_event_row"] = (df[event_col] == 1) & (df["_cum_event"].groupby(df[id_col]).cumsum() == 1)

def pick_row(g):
    event_rows = g[g[event_col] == 1]
    if len(event_rows) > 0:
        return event_rows.iloc[0]   # first time it hit poor
    else:
        return g.iloc[-1]           # last observed survey (censored)

bridge_level = df.groupby(id_col, group_keys=False).apply(pick_row).reset_index(drop=True)
bridge_level = bridge_level.drop(columns=["_cum_event", "_is_first_event_row"])

# ---- 5. Build final time/event columns ----
bridge_level["time"] = bridge_level[age_col].astype(float)
bridge_level["event"] = bridge_level[event_col].astype(bool)  # sksurv wants boolean event

# ---- 6. Drop rows with unusable time values ----
n_before = len(bridge_level)
bridge_level = bridge_level[bridge_level["time"] > 0]
n_after = len(bridge_level)
print(f"Dropped {n_before - n_after} bridges with non-positive time (bad bridge_age).")

# ---- 7. Quick sanity checks ----
print(f"\nFinal bridge-level dataset: {len(bridge_level)} bridges")
print(f"Events (transitioned to poor): {bridge_level['event'].sum()}")
print(f"Censored (never observed transitioning): {(~bridge_level['event']).sum()}")
print(f"Time range: {bridge_level['time'].min():.1f} to {bridge_level['time'].max():.1f} years")

# ---- 8. Build sksurv structured array target ----
from sksurv.util import Surv
y = Surv.from_dataframe("event", "time", bridge_level)

print("\nSample of final time/event columns:")
print(bridge_level[[id_col, year_col, age_col, "time", "event"]].head(10))

# ---- 9. Save outputs ----
bridge_level.to_csv("RSF_data_a.csv", index=False)
print("\nSaved to RSF_data_a.csv")

# X (features) would then be bridge_level.drop(columns=[id_col, year_col, "time", event_col, "event", ...other leakage cols])
# y is the structured array above, ready for RandomSurvivalForest.fit(X, y)
```

```python
"""
OPTION B: Bridge-year rows treated as pseudo-subjects for sksurv RSF.

This is a SECONDARY / SENSITIVITY analysis, not the primary model. It keeps
every bridge-year row (and therefore every time-varying climate/maintenance
reading) instead of collapsing each bridge to a single row, at the cost of
violating the standard survival-analysis independence assumption: multiple
rows from the same physical bridge are treated as if they were different
subjects.

Because sksurv's RandomSurvivalForest requires one (event, time) pair per
ROW (it has no native panel/counting-process support), this script:

1. Keeps each bridge-year row up through and including the row where the
   bridge first transitions to "poor" (event == 1), then drops all later
   rows for that bridge (it's already failed, shouldn't still appear "at risk").
2. Sets `time` = bridge_age for that row (NOT a start/stop interval --
   sksurv doesn't support that; this is each row's age at observation,
   matching what RSF expects).
3. Sets `event` = the row's own event flag (1 only on the bridge's final
   kept row if it transitioned, 0 on all earlier rows for that bridge and
   for bridges that are fully censored).
4. Builds a `bridge_group` column (== STRUCTURE_NUMBER_008) that you MUST
   use with GroupKFold / GroupShuffleSplit in any train/test split or
   cross-validation. Splitting randomly by row instead of by group is the
   single most common way this approach silently inflates performance --
   a bridge's age-10 row could land in train and its age-20 row in test,
   meaning the model has effectively already seen that bridge.

Use this alongside the Option A (one-row-per-bridge) model as a comparison /
robustness check, not as a replacement for it.
"""

import pandas as pd
import numpy as np

# ---- 1. Load your panel (bridge-year) data ----
# df = pd.read_csv("your_bridge_panel.csv")
df = pd.read_csv("massachusetts_bridges_rsf_ready.csv")  # update path

id_col = "STRUCTURE_NUMBER_008"
year_col = "SURVEY_YEAR"
age_col = "bridge_age"
age_flag_col = "bridge_age_negative_flag"
event_col = "event"

# ---- 2. Check bridge_age data quality ----
n_bad_age = (df[age_flag_col] == 1).sum() if age_flag_col in df.columns else (df[age_col] < 0).sum()
print(f"Rows with negative/bad bridge_age: {n_bad_age} of {len(df)}")
bad_age_structures = df.loc[df[age_col] < 0, id_col].unique()
if len(bad_age_structures) > 0:
    print(f"{len(bad_age_structures)} structures have at least one negative bridge_age row -- "
          f"inspect/fix YEAR_BUILT_027 for these before modeling.")
    print(bad_age_structures[:10])

# ---- 3. Sort within each bridge by survey year (required for cumulative logic) ----
df = df.sort_values([id_col, year_col]).reset_index(drop=True)

# ---- 4. Drop bridge-year rows AFTER a bridge's first event ----
# Once a bridge hits event == 1, it shouldn't contribute further "still at
# risk" pseudo-subject rows.
df["_cum_event"] = df.groupby(id_col)[event_col].cummax()
df["_had_prior_event"] = df.groupby(id_col)["_cum_event"].shift(1).fillna(0)
panel = df[df["_had_prior_event"] == 0].copy()
panel = panel.drop(columns=["_cum_event", "_had_prior_event"])

# ---- 5. Build time/event columns for sksurv ----
panel["time"] = panel[age_col].astype(float)
panel["event"] = panel[event_col].astype(bool)

# ---- 6. Drop rows with unusable time values ----
n_before = len(panel)
panel = panel[panel["time"] > 0]
n_after = len(panel)
print(f"Dropped {n_before - n_after} rows with non-positive time (bad bridge_age).")

# ---- 7. Drop degenerate duplicate (bridge, age) rows ----
# If a bridge has two surveys landing on the same bridge_age (e.g. two
# inspections in one year), sksurv doesn't care, but it's worth knowing about.
dup_mask = panel.duplicated(subset=[id_col, "time"], keep=False)
if dup_mask.sum() > 0:
    print(f"Note: {dup_mask.sum()} rows share the same (bridge, time) pair -- "
          f"likely duplicate-year inspections. Not dropped automatically, inspect if this is large.")

# ---- 8. Build the grouping column for GroupKFold / GroupShuffleSplit ----
panel["bridge_group"] = panel[id_col]

# ---- 9. Sanity checks ----
n_bridges = panel[id_col].nunique()
n_rows = len(panel)
n_events = panel["event"].sum()
print(f"\nFinal pseudo-subject panel: {n_rows} rows from {n_bridges} unique bridges")
print(f"Avg rows per bridge: {n_rows / n_bridges:.2f}")
print(f"Event rows (transitioned to poor): {n_events}")
print(f"Censored rows: {n_rows - n_events}")
print(f"Time range: {panel['time'].min():.1f} to {panel['time'].max():.1f} years")

# ---- 10. Build sksurv structured array target ----
from sksurv.util import Surv
y = Surv.from_dataframe("event", "time", panel)

print("\nSample of final time/event/group columns:")
print(panel[[id_col, year_col, age_col, "time", "event", "bridge_group"]].head(10))

# ---- 11. Save outputs ----
panel.to_csv("RSF_bridge_panel_pseudo_subjects.csv", index=False)
print("\nSaved to RSF_bridge_panel_pseudo_subjects.csv")

# ---- 12. REQUIRED: grouped train/test split example ----
# Never use a plain random train_test_split here -- it will leak bridges
# across train/test and inflate your reported performance.
#
# from sklearn.model_selection import GroupShuffleSplit
# X = panel.drop(columns=[id_col, year_col, "time", event_col, "event", "bridge_group", ...other leakage cols])
# groups = panel["bridge_group"]
#
# gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
# train_idx, test_idx = next(gss.split(X, y, groups=groups))
# X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
# y_train, y_test = y[train_idx], y[test_idx]
#
# For cross-validation during model selection, use GroupKFold instead of KFold:
# from sklearn.model_selection import GroupKFold
# gkf = GroupKFold(n_splits=5)
# for train_idx, val_idx in gkf.split(X, y, groups=groups):
#     ...
```
